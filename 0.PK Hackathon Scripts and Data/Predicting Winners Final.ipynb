# Team Stats
Incorporating the previous season wins, losses and rank of team1 and team2 as features

```python
lis=[]
for i in range(0,len(files)):
    if 'standings' in files[i]:
        lis.append(files[i])
lis
```

```python
f2=pd.DataFrame()
for f in lis:
    s=pd.read_csv(path+"\\"+f)
    s['Season']=f[0:8]
    f2=pd.concat([f2,s])
f2.head()
```

```python
#Cleaning/Structuring the data
f2=f2[f2.Team!="teamteam"]
f2.Season=f2.Season.str.title()
f2['Rank']=f2.Rank.replace({'1q':1,'2q':2,'3q':3})
```

```python
final_df['Team1_previous_seasons_wins']=0
final_df['Team1_previous_seasons_losses']=0
final_df['Team1_previous_seasons_Rank']=0
#final_df['Team1_previous_seasons_Points']=0

m=list(final_df.Team2.value_counts().index)
season_i=['Season 2','Season 3','Season 4','Season 5','Season 6','Season 7']
season_j=['Season 1','Season 2','Season 3','Season 4','Season 5','Season 6']

f2.Won=f2.Won.astype('int64')
f2.Lost=f2.Lost.astype('int64')
f2.Rank=f2.Rank.astype('int64')
f2.Points=f2.Points.astype('int64')

#For Team1
for team in m:
    for season in range(0,len(season_i)):
        i=((final_df['Team1']==team) & (final_df['Season']==season_i[season]))
        j=((f2['Team']==team) & (f2['Season']==season_j[season]))
        
        #Correctly adding wins and losses of the previous season to the df
        if f2[j]['Won'].empty: #In some cases where the team is newly formed the result is an empty series
            final_df['Team1_previous_seasons_wins'][i]=0
        elif f2[j]['Lost'].empty:
            final_df['Team1_previous_seasons_losses'][i]=0
        elif f2[j]['Rank'].empty:
            final_df['Team1_previous_seasons_Rank'][i]=0
        #elif f2[j]['Points'].empty:
            #final_df['Team1_previous_seasons_Points'][i]=0    
        else:    
            final_df['Team1_previous_seasons_wins'][i]=f2[j]['Won'].values
            final_df['Team1_previous_seasons_losses'][i]=f2[j]['Lost'].values
            final_df['Team1_previous_seasons_Rank'][i]=f2[j]['Rank'].values
            #final_df['Team1_previous_seasons_Points'][i]=f2[j]['Points'].values
```

```python
final_df['Team2_previous_seasons_wins']=0
final_df['Team2_previous_seasons_losses']=0
final_df['Team2_previous_seasons_Rank']=0
#final_df['Team2_previous_seasons_Points']=0

m=list(final_df.Team2.value_counts().index)
season_i=['Season 2','Season 3','Season 4','Season 5','Season 6','Season 7']
season_j=['Season 1','Season 2','Season 3','Season 4','Season 5','Season 6']

for team in m:
    for season in range(0,len(season_i)):
        i=((final_df['Team2']==team) & (final_df['Season']==season_i[season]))
        j=((f2['Team']==team) & (f2['Season']==season_j[season]))
        
        #Correctly adding wins and losses of the previous season to the df
        if f2[j]['Won'].empty: #In some cases where the team is newly formed the result is an empty series
            final_df['Team2_previous_seasons_wins'][i]=0
        elif f2[j]['Lost'].empty:
            final_df['Team2_previous_seasons_losses'][i]=0
        elif f2[j]['Rank'].empty:
            final_df['Team2_previous_seasons_Rank'][i]=0    
        #elif f2[j]['Points'].empty:
            #final_df['Team2_previous_seasons_Points'][i]=0    
        else:    
            final_df['Team2_previous_seasons_wins'][i]=f2[j]['Won'].values
            final_df['Team2_previous_seasons_losses'][i]=f2[j]['Lost'].values
            final_df['Team2_previous_seasons_Rank'][i]=f2[j]['Rank'].values
            #final_df['Team2_previous_seasons_Points'][i]=f2[j]['Points'].values
```

# Eliminator matches

```python
#A simple function to predict the winners of the elimination matches
def predict_winners(team1,team2,team1_w,team1_l,team1_r,team2_w,team2_l,team2_r,predictors,l):
    predictors['Team1']=team1
    predictors['Team2']=team2
    predictors['Team1_previous_seasons_wins']=team1_w
    predictors['Team1_previous_seasons_losses']=team1_l
    predictors['Team1_previous_seasons_Rank']=team1_r
    predictors['Team2_previous_seasons_wins']=team2_w
    predictors['Team2_previous_seasons_losses']=team2_l
    predictors['Team2_previous_seasons_Rank']=team2_r

    predictors_sf=predictors.drop(['Winning Team','match_status','Team1','Team2','match_number','Team1_venue'],axis='columns')
    predictors_sf=pd.get_dummies(predictors_sf)
    #We know that no team plays at home in the eliminator matches so there will be no home team venue
    predictors_sf['Team1_venue_Home']=0
    lis=[]
    for i in range(0,len(l)):
        if 'Season' in l[i]:
            lis.append(l[i])
    #lis
    for i in lis:
        predictors_sf[i]=0
    predictors_sf['Season_Season 7']=1    
    final_predictions_sf=clf3.predict(predictors_sf)
    return(final_predictions_sf)
    
```

```python
#The teams which qualified for the semifinal matches (as determined by the predicted standings)
team1=['Bengaluru Bulls','U.P. Yoddha']
team2=['Haryana Steelers','U Mumba']
team1_w=[13,8]
team1_l=[7,10]
team1_r=[1,3]
team2_w=[6,15]
team2_l=[14,5]
team2_r=[6,2]
predictors=sf_eliminators
l=list(predictors_test.columns)
```

```python
winners=predict_winners(team1,team2,team1_w,team1_l,team1_r,team2_w,team2_l,team2_r,predictors,l)
winners
```
